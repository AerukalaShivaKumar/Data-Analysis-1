# 2.What is the Average price of vehicle by fuel type and gearbox
## type.Give a plot

```python
plot = df[['fuelType','gearbox',"price"]]
```

```python
plot
```

```python
average_prices = df.groupby(['fuelType','gearbox'])['price'].mean().reset_index()
average_price =average_prices[average_prices['fuelType']!="andere"]
average_price
```

```python
average_price
```

```python
sns.barplot(x=average_price['fuelType'],y=average_price['price'],hue=average_price['gearbox'])
plt.xticks(rotation=45)
plt.show()
```

```python
average_price_1=average_prices[average_prices['fuelType']=="andere"]
average_price_1
```

```python
sns.barplot(x=average_price_1['fuelType'],y=average_price_1['price'],hue=average_price_1['gearbox'])
plt.xticks(rotation=45)
plt.show()
```

# 4.What is the Average price of a vehicle by brand as well as
## vehicle type.Use heatmap to explain this

```python
df1 = df[['brand','vehicleType','price']]
```

```python
sns.set_theme(palette='pastel')
plt.figure(figsize=(30,20))
heat_table = df1.pivot_table(values='price',index='brand',columns='vehicleType',aggfunc=np.median)
sns.heatmap(heat_table,cmap='Greens',annot=True)
plt.show()
```
